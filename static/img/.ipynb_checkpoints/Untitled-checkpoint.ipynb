```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def get_color_mask(mask, color, input_shape):
    bin_mask = np.zeros(input_shape)
    color_mask = np.zeros(input_shape)
    for i in range(3):
        color_mask[:,:,i] = mask*color[i]
        bin_mask[:,:,i] = mask
    return np.uint8(color_mask), np.uint8(bin_mask)
```

```python
image = cv2.imread('img5.png')[...,::-1]
ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
mask = np.zeros((512, 512, 3))
for j in range(512):
    for i in range(512):
        if thresh1[i, j, 0] == 255 and thresh1[i, j, 1] == 0 and thresh1[i, j, 2] == 0:
            mask[i, j, :] = [1, 1, 1]
        else:
            continue 

color_mask, mask = get_color_mask(mask[:,:,0], [255/255., 237/237., 0], (512, 512, 3))
image = image
background = np.uint8(image)*np.uint8(1-mask)
hair = np.uint8(image)*np.uint8(mask)
hair[hair>128] = 255
hair[hair<=128] = 0
color_mask[color_mask<=228] = 0
new_image = np.uint8(np.clip(hair*0.8 + color_mask*0.2, 0, 255))
new_image[new_image<200] = 0
```

```python
plt.imshow(new_image+background)
```

```python
cv2.imwrite('img_7.png', new_image[...,::-1])
```
